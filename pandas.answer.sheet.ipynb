```python
import pandas as pd
import numpy as np

## 1 使用List构造Series

data = pd.Series([1, 2, 3, 4, 5])

## 2 使用Dict构造Series
dict = pd.Series({ '语文': 98, '数学': 99, '英语': 100 })
dict

## 3 Series转换成List
dict.to_list()



## 5 借助 Numpy 创建Series

data2 = pd.Series(np.arange(10), dtype="float")

data2

## 6 转换Series的数据类型

data2.map(int)

data2.astype(int)

## 7 给Series添加新的元素

data2.append(pd.Series(['a', 'b', 'c']))

## 8 用reset index将Series转换成df

data2.reset_index()


## 4 将Series转换成DataFrame

df = pd.DataFrame(
 ['add', 'bas', 'cwq', 'dxx', 'ewqq'], 
  columns=['A'],
  index=['a', 'b', 'c', 'd', 'e']
)

df
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
  pd.date_range('2023-10-1', '2023-11-1', closed='left'),
  columns=["days"]
)


df['dayofyear'] = df["days"].dt.dayofyear

df

df.set_index('days', inplace=False)
```
